```python
def numPrimeArrangements(n):
    def Fact(n):
        if n==1:
            return 1
        else:
            return n*Fact(n-1)
    def isprime(n):
        for i in range(2,int(math.sqrt(n))+1):
            if n%i==0:
                return False
        return True
    def Prime(n):
        count=0
        for i in range(2,n+1):
            if isprime(i):
                count+=1
        return count
    if n<3:
        return 1
    a = Prime(n)
    b = n - a
    return (Fact(a)*Fact(b)%1000000007+1000000007)%1000000007
```

```python
def minstep(n): 
    def dfs(n,dp):
        if n==1:
            return 0
        if dp[n]!=0:
            return dp[n]
        if n%2==0:
            dp[n] = 1+min(dfs(n//2,dp),dfs(n-1,dp))
        elif n%3==0:
            dp[n] = 1+min(dfs(n//3,dp),dfs(n-1,dp))
        else:
            dp[n] = 1+dfs(n-1,dp)
        return dp[n]
    dp = [0 for i in range(n+1)]
    return dfs(n,dp)
minstep(5)
```

```python
def isPerfectSquare(num):
    l = 1
    h = num
    
    while l<=h:
        mid = l+(h-l)//2
        if mid*mid == num:
            return True
        if mid*mid > num:
            h = mid - 1
        else:
            l = mid + 1
    return False
isPerfectSquare(1)
```

```python
def FriendsPairing(n):
    def Top(n,dp):
        if n == 0:
            return 0
        if n <= 2:
            return n
        if dp[n]!=0:
            return dp[n]
        else:
            dp[n] = FriendsPairing(n-1) + (n-1)*FriendsPairing(n-2)
        return dp[n]
    dp = [ 0 for i in range(n+1)]
    return Top(n,dp)
FriendsPairing(5)
```

```python
def findJudge(N,trust):
    inDegree = [0]*(N+1)
    outDegree = [0]*(N+1)
    for i in trust:
        inDegree[i[1]] += 1
        outDegree[i[0]] += 1
    for i in range(1,N+1):
        if inDegree[i] == N-1 and outDegree[i] == 0:
            return i
    return -1
findJudge(3,[[1,3],[2,3],[3,1]])
```

```python
def floodFill(image,sr,sc,newColor):
    def dfs(image,i,j,newColor,old):
        if i<0 or j<0 or i>len(image)-1 or j>len(image[0])-1 or image[i][j] != old:
            return 
        image[i][j] = newColor
        dfs(image,i+1,j,newColor,old)
        dfs(image,i,j+1,newColor,old)
        dfs(image,i-1,j,newColor,old)
        dfs(image,i,j-1,newColor,old)
        return image
    old = image[sr][sc]
    if newColor == old:
        return image
    return dfs(image,sr,sc,newColor,old)
floodFill([[0,0,0],[0,1,1]],1,1,1)
```

```python
def AlicPower(nums):
    negative = [ 0 for i in range(len(nums))]
    positive = [ 0 for i in range(len(nums))]

    positive[0] = negative[0] = nums[0]
    max_so_far = nums[0]
    for i in range(1,len(nums)):
        positive[i] = max(positive[i-1]*nums[i],negative[i-1]*nums[i],nums[i])
        negative[i] = min(positive[i-1]*nums[i],negative[i-1]*nums[i],nums[i])
        max_so_far = max(positive[i],max_so_far)
    return max_so_far
AlicPower([10,1,2,-13,0,7,-18,2])
```

```python
def MaxiBounties(s):
    stack = [-1]
    maxi = 0
    for i in range(len(s)):
        if s[i] == '(':
            stack.append(i)
        else:
            if stack[-1] == -1 or s[stack[-1]]!="(":
                stack.append(i)
            else:
                stack.pop()
                maxi = max(maxi,i-stack[-1])
    return maxi        
    
MaxiBounties("()(()))())()")
```

```python
from collections import defaultdict
def kClosest(points,K):
    maps = defaultdict(list)
    for i,j in points:
        ans = (i*i) + (j*j)
        maps[ans].append([i,j])
    res = []
    temp = sorted(maps)
    for i in temp:
        for j in maps[i]:
            if K:
                res.append(j)
            else:
                return res
            K-=1
    return res
kClosest([[3,3],[5,-1],[-2,4]],2)
```

```python
def MagicValen(a1,a2):
    def Main(b,g,dp,i,j,n,m):
        if i == n:
            return 0
        if j == m:
            return float('inf')
        if dp[i][j]!=-1:
            return dp[i][j]
        op1 = abs(b[i]-g[j]) + Main(b,g,dp,i+1,j+1,n,m)
        op2 = Main(b,g,dp,i,j+1,n,m)
        dp[i][j] = min(op1,op2)
        return dp[i][j]
    n = len(a1)
    m = len(a2)
    dp = [ [ -1 for i in range(m+1)] for j in range(n+1)]
    return Main(a1,a2,dp,0,0,n,m)
MagicValen([4,5],[1,2,3,4,6])
```

```python
def ModivsTrump(arr):
    def AmitShah(arr,i,j,dp):
        if i > j:
            return 0
        if dp[i][j]!=-1:
            return dp[i][j]
        Modi,Trump = 0,0
        Modi = arr[i] + min(AmitShah(arr,i+2,j,dp),AmitShah(arr,i+1,j-1,dp))
        Trump = arr[j] + min(AmitShah(arr,i+1,j-1,dp),AmitShah(arr,i,j-2,dp))
        dp[i][j] = max(Modi,Trump)
        return dp[i][j]
    dp = [ [-1 for j in range(len(arr)+1)] for i in range(len(arr)+1)]
    return AmitShah(arr,0,len(arr)-1,dp)
ModivsTrump([2,4,7,6])
```

```python
def EggDropping(n,k):
    def Helper(n,k):
        if n==0 or n == 1:
            return n
        if k == 1:
            return n
        ans = float('inf')
        for i in range(1,n+1):
            res = max(Helper(i-1,k-1),Helper(n-i,k))
            ans = min(ans,res)
        return ans+1
    return Helper(n,k)
EggDropping(10,2)
```

```python
from collections import defaultdict
import heapq
def findCheapestPrice(n,flights,src,dst,K):
    maps = defaultdict(dict)
    for i,j,z in flights:
        maps[i][j] = z
    heap = [ (0,src,K+1) ]
    while heap:
        p,i,z = heapq.heappop(heap)
        if i == dst:
            return p
        if z>0:
            for j in maps[i]:
                heapq.heappush(heap,(p+maps[i][j],j,z-1))
    return -1
findCheapestPrice(3,[[0,1,100],[1,2,100],[0,2,500]],0,2,1)
```
